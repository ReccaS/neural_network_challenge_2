## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with 'Attrition' and 'Department' columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = [
    'Age',
    'BusinessTravel',
    'OverTime',
    'JobRole',
    'JobSatisfaction',
    'WorkLifeBalance',
    'YearsAtCompany',
    'RelationshipSatisfaction',
    'EnvironmentSatisfaction',
    'Education',
    'PerformanceRating',
    'MaritalStatus',
    'TotalWorkingYears',
    'JobInvolvement'
]


# Create X_df using your selected columns
X_df = attrition_df[selected_columns]

# Show the data types for X_df
print("Data types of X_df columns:")
print(X_df.dtypes)
#print(X_df_numeric.dtypes)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_df,       # Features
    y_df,          # Target variable
    test_size=0.2,           # 20% for testing
    random_state=42,         # For reproducibility
    stratify=y_df               # To maintain class distribution
)

# Verify the shapes of the splits
print("Shapes of the datasets:")
print(f"X_train: {X_train.shape}")  # Expected: (3448, 15)
print(f"X_test: {X_test.shape}")    # Expected: (861, 15)
print(f"y_train: {y_train.shape}")  # Expected: (3448,)
print(f"y_test: {y_test.shape}")    # Expected: (861,)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

# Convert categorical variables to numeric using One-Hot Encoding
categorical_cols = ['BusinessTravel', 'OverTime', 'JobRole', 'MaritalStatus', 'Education']
X_train_numeric = pd.get_dummies(X_train, columns=categorical_cols)
X_test_numeric = pd.get_dummies(X_test, columns=categorical_cols)

# Show the first few rows of the updated X_df_numeric
print("First 5 rows of X_df after encoding categorical variables:")
display(X_train_numeric.head())

# Show the data types for X_df_numeric
print("Data types of X_df columns:")
print(X_train_numeric.dtypes)
```

```python
# Create a StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
display(X_train_numeric.head())
scaler.fit(X_train_numeric)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train_numeric)
X_test_scaled = scaler.transform(X_test_numeric)

# Convert the scaled arrays back to DataFrames with original column names
X_train = pd.DataFrame(X_train_scaled, columns=X_train_numeric.columns, index=X_train.index)
X_test = pd.DataFrame(X_test_scaled, columns=X_test_numeric.columns, index=X_test.index)

# (Optional) Display the first few rows of the scaled training data
print("First 5 rows of X_train after scaling:")
display(X_train.head())

```

```python
# Create a OneHotEncoder for the Department column
department_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
department_encoder.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data

# Transform the 'Department' column in training and testing data
y_train_dept_encoded = department_encoder.transform(y_train[['Department']])
y_test_dept_encoded = department_encoder.transform(y_test[['Department']])
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
attrition_encoder.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data

# Transform the 'Attrition' column in training and testing data
y_train_attrition_encoded = attrition_encoder.transform(y_train[['Attrition']])
y_test_attrition_encoded = attrition_encoder.transform(y_test[['Attrition']])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_dim = X_train.shape[1] # Number of columns/features

# Create the input layer
input_layer = layers.Input(shape=(input_dim,), name="input")

# Create at least two shared layers
shared_layer_1 = layers.Dense(64, activation="relu", name="shared1")(input_layer)
shared_layer_2 = layers.Dense(128, activation="relu", name="shared2")(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_hidden_layer = layers.Dense(32, activation="relu", name="department_hidden")(shared_layer_2)

# Create the output layer
dept_output_layer = layers.Dense(2, activation="softmax", name="department_output")(dept_hidden_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden_layer = layers.Dense(32, activation="relu", name="attrition_hidden")(shared_layer_2)

# Create the output layer
attrition_output_layer = layers.Dense(3, activation="softmax", name="attrition_output")(attrition_hidden_layer)
```

```python
# Create the model
#model = Model(inputs=input_layer, outputs=[dept_output_layer, attrition_output_layer])
model = Model(inputs=input_layer, outputs=[dept_output_layer, attrition_output_layer])
# Compile the model
model.compile(
    optimizer="adam",
    loss={
        "department_output": "categorical_crossentropy",
        "attrition_output": "categorical_crossentropy"  # Changed from 'binary_crossentropy'
    },
    metrics={
        "department_output": "accuracy",
        "attrition_output": "accuracy"
    }
)

# Summarize the model
model.summary()
```

```python
print("Shapes of the datasets:")
print(f"X_train: {X_train.shape}")
print(f"y_train_dept_encoded: {y_train_dept_encoded.shape}")
print(f"y_train_attrition_encoded: {y_train_attrition_encoded.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_test_dept_encoded: {y_test_dept_encoded.shape}")
print(f"y_test_attrition_encoded: {y_test_attrition_encoded.shape}")

print(f"Shape of y_train_dept_encoded: {y_train_dept_encoded.shape}")
print(f"Shape of y_train_attrition_encoded: {y_train_attrition_encoded.shape}")
print("Model output names:", model.output_names)
```

```python
print("y_train_dept_encoded (first 5 rows):\n", y_train_dept_encoded[:5])
print("y_train_attrition_encoded (first 5 rows):\n", y_train_attrition_encoded[:5])
print("y_test_dept_encoded (first 5 rows):\n", y_test_dept_encoded[:5])
print("y_test_attrition_encoded (first 5 rows):\n", y_test_attrition_encoded[:5])

history = model.fit(
    X_train,
    {
        "department_output": y_train_dept_encoded,
        "attrition_output": y_train_attrition_encoded
    },
    epochs=20,
    batch_size=32,
    validation_data=(
        X_test,
        {
            "department_output": y_test_dept_encoded,
            "attrition_output": y_test_attrition_encoded
        }
    )
)
```

```python
# Evaluate the model with the testing data

test_results = model.evaluate(

    X_test,

    {

        "department_output": y_test_dept_encoded,

        "attrition_output": y_test_attrition_encoded

    },

    verbose=1  # Set to 1 to see the progress bar

)

print(test_results)
```

```python
# Print the accuracy for both department and attrition

# Retrieve metric names

metric_names = model.metrics_names

print("Metrics Names:", model.metrics_names)

 

# Combine metric names and results into a dictionary

results_dict = dict(zip(metric_names, test_results))

print(results_dict)

# Extract accuracy values

department_accuracy = results_dict['department_output_accuracy']

attrition_accuracy = results_dict['attrition_output_accuracy']

 

# Print the accuracies

print(f"Department predictions accuracy: {department_accuracy}")

print(f"Attrition predictions accuracy: {attrition_accuracy}")
```
